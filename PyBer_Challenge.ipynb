### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Challenge Deliverable 1. Generate a Ride-Sharing DataFrame by City Type

```python
#  1. Get the total rides for each city type
city_type_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
city_type_ride_count
```

```python
# 2. Get the total drivers for each city type
city_type_driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
city_type_driver_count
```

```python
#  3. Get the total amount of fares for each city type
city_type_fare_amount = pyber_data_df.groupby(["type"]).sum()["fare"]
city_type_fare_amount
```

```python
#  4. Get the average fare per ride for each city type. 
city_type_fare_avg = pyber_data_df.groupby(["type"]).mean()["fare"]
city_type_fare_avg
```

```python
# 5. Get the average fare per driver for each city type. 
city_type_fare_bydriver = city_type_fare_amount/city_type_driver_count
city_type_fare_bydriver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summ_df = pd.DataFrame()
pyber_summ_df["Total Rides"] = city_type_ride_count
pyber_summ_df["Total Drivers"] = city_type_driver_count
pyber_summ_df["Total Fares"] = city_type_fare_amount
pyber_summ_df["Average Fare per Ride"] = city_type_fare_avg
pyber_summ_df["Average Fare per Driver"] = city_type_fare_bydriver
pyber_summ_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summ_df.index.name = None
pyber_summ_df.head()
```

```python
#  8. Format the columns.
pyber_summ_df["Total Fares"] = pyber_summ_df["Total Fares"].map("${:,.2f}".format)
pyber_summ_df["Average Fare per Ride"] = pyber_summ_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summ_df["Average Fare per Driver"] = pyber_summ_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_summ_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# Print the merged DataFrame for reference.
pyber_data_df
pyber_data_df['date'] = pd.to_datetime(pyber_data_df['date']).dt.date
```

```python
# 1. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fares_by_typedate_df = pyber_data_df[["type","date","fare"]].groupby(["type","date"]).sum()
fares_by_typedate_df
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fares_by_typedate_df = fares_by_typedate_df.reset_index()
fares_by_typedate_df
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_df = pd.pivot_table(fares_by_typedate_df,index=["date"],columns = ["type"],values = ["fare"])
pivot_df.index.name = None
pivot_df
```

```python
# 4. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
start_date = pd.to_datetime('2019-01-01')
end_date = pd.to_datetime('2019-04-29')
fares_pivot_df = pivot_df.loc[start_date:end_date]
fares_pivot_df
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
fares_pivot_df.index = pd.to_datetime(fares_pivot_df.index)
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
fares_pivot_df.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fares_pivot_df = fares_pivot_df.resample('W').sum()
```
