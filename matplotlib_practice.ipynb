```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis, label = 'Boston', marker = "*", color="blue", linewidth = 2) # have to use legend to show label
plt.xlabel('Month')
plt.ylabel('Fares')
plt.title('PyBer Fare by Month')
plt.ylim(0,45)
plt.xlim(0,5)
plt.legend()
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots() # returns a tuple - figure and axes objects. They are assigned to two variables
ax.plot(x_axis,y_axis, label = 'Boston')
```

```python
#alternatively we can assign variable individually
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis,color="green", marker="D",linewidth = 2,label="Boston")
ax.legend()
ax.set_title("Title")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_ylim(0,45)
ax.grid()
plt.show()
fig.savefig("fig.pdf")
```

```python
# Scatter Plots
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis,'o') # add 'o' for scatter plot
```

```python
plt.scatter(y_axis,x_axis, color = "Red", label = 'Chicago')
plt.legend()
plt.title("T")
plt.xlabel("x")
plt.ylabel("y")
plt.gca().invert_yaxis()
plt.xlim(15,40)
plt.show()
```

```python
# Bubble
larger = []
for data in y_axis:
    larger+=([data*3])
    #print(larger)
plt.scatter(x_axis, y_axis, s=larger)
```

```python
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis if i > 0]) # if is optional. this is called list comprehension
```

```python
# scatter with OO Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# bubble with OO Interface
fig, ax = plt.subplots()
ax.scatter(y_axis,x_axis, s=[i*5 for i in y_axis], color="skyblue", alpha = .2, label = "legend")
ax.set_title("Title")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.invert_yaxis() # no need for gca in OO method
ax.set_xlim(0,50)
#ax.gca().
ax.legend()
```

```python
# pie chart
plt.pie(y_axis, labels=x_axis, autopct = "%1.1f%%")
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize= (8,8)) # this is before plt.pie
explode = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
        labels=x_axis, 
        autopct = "%.1f%%",
        explode=explode,
       colors=colors)
plt.show()
```

```python
# error bars
%matplotlib inline
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis,y_axis,yerr=stdev,capsize=3)
```

```python
plt.bar(x_axis,y_axis,yerr=stdev,capsize=3)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
# major ticks
import numpy as np
plt.barh(x_axis,y_axis)
plt.xticks(np.arange(0,51, step=5.0))
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
plt.show()
```

```python
# minor ticks
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
